```python
array1 = np.random.randint(1, 11, size=5)
array2 = np.random.randint(1, 11, size=5)

print("Перший масив:", array1)
print("Другий масив:", array2)

addition = array1 + array2
subtraction = array1 - array2
multiplication = array1 * array2

print("Результат додавання:", addition)
print("Результат віднімання:", subtraction)
print("Результат множення:", multiplication)
```

```python
matrix1 = np.random.randint(1, 11, size=(2, 2))
matrix2 = np.random.randint(1, 11, size=(2, 3))

print("Перша матриця (2x2):")
print(matrix1)
print("Друга матриця (2x3):")
print(matrix2)

result = np.dot(matrix1, matrix2)

print("Результат множення матриць:")
print(result)
```

```python
matrix = np.random.randint(1, 11, size=(3, 3))

print("Початкова матриця (3x3):")
print(matrix)

try:
    inverse_matrix = np.linalg.inv(matrix)
    print("Обернена матриця:")
    print(inverse_matrix)
except np.linalg.LinAlgError:
    print("Ця матриця є сингулярною (неможливо знайти обернену матрицю).")
```

```python
matrix = np.random.rand(4, 4)

print("Початкова матриця (4x4):")
print(matrix)

transposed_matrix = np.transpose(matrix)

print("Транспонована матриця:")
print(transposed_matrix)
```

```python
matrix = np.random.randint(1, 11, size=(3, 4))

vector = np.random.randint(1, 11, size=4)

print("Матриця (3x4):")
print(matrix)
print("Вектор (розмір 4):")
print(vector)

result = np.dot(matrix, vector)

print("Результат перемноження матриці на вектор:")
print(result)
```

```python
matrix1 = np.random.randint(1, 11, size=(2, 2))
matrix2 = np.random.randint(1, 11, size=(2, 2))

print("Перша матриця (2x2):")
print(matrix1)
print("Друга матриця (2x2):")
print(matrix2)

elementwise_product = np.multiply(matrix1, matrix2)

print("Результат поелементного множення матриць:")
print(elementwise_product)
```

```python
matrix1 = np.random.randint(1, 11, size=(4, 4))
matrix2 = np.random.randint(1, 11, size=(4, 4))

print("Перша матриця (4x4):")
print(matrix1)
print("Друга матриця (4x4):")
print(matrix2)

difference = matrix1 - matrix2

print("Різниця матриць:")
print(difference)
```

```python
matrix = np.random.rand(3, 3)

print("Матриця (3x3):")
print(matrix)

row_sums = np.sum(matrix, axis=1)
column_vector = row_sums.reshape(3, 1)

print("Вектор-стовпчик із сумами елементів кожного рядка:")
print(column_vector)
```

```python
matrix = np.random.randint(1, 11, size=(3, 4))

print("Початкова матриця (3x4):")
print(matrix)

squared_matrix = np.square(matrix)

print("Матриця з квадратами чисел:")
print(squared_matrix)
```

```python
vector = np.random.randint(1, 51, size=4)

print("Початковий вектор:")
print(vector)

sqrt_vector = np.sqrt(vector)

print("Вектор з квадратними коренями:")
print(sqrt_vector)
```
